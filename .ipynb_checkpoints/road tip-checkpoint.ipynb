```python
distance = float(input("Enter total distance"))
kpl = float(input("Enter km per liter"))
price = float(input("Enter fuel price liter"))

liters = distance / kpl
cost =liters * price

cost = round(cost,2)
liters = round(liters,2)

print("\n Road Trip Summary")
print("----------------------")
print("Distance:", distance, "km")
print("Car Efficiency:", kpl, "km/L")
print("Fuel Price: $", price)
print("\nFuel Needed:", liters, "liters")
print("Total Cost: $", cost)


```
